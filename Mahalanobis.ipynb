```python
#08_22_2021_mahalanobis_v3.ipynb

def mahalanobis(lat, data, x):
    
    
    ## get data for a single latitude
    data_1L = data[:,lat,:]
    x_1L = x[:,lat,:]
    
    if np.isnan(x_1L[0,3])==True:
        
        nn = np.zeros(100)
        nn[:] = np.nan

        return nn, nn
    
    elif np.isnan(data_1L[0,3])==True:
        
        nn = np.zeros(100)
        nn[:] = np.nan

        return nn, nn
    
    else:
    

        ## get a1 and b1 (both ensembles)
        da1, db1 = data_1L[:,8],  data_1L[:,9]
        xa1, xb1 = x_1L[:,8],  x_1L[:,9]

        ## put into arrays, in order to compute mahalabobis
        data_arr=np.vstack((da1,db1))
        x_arr=np.vstack((xa1,xb1))

        ## get the mean of data 
        data_mean=np.mean(data_arr,axis=1)

        ## subtract mean of data from each value in x
        x_minus_mu = x_arr - data_mean.reshape((2,1))

        ## this is to have the dimensions line up correctly
        x_minus_mu = x_minus_mu.T

        ## compute the invers covariance matrix of data array
        cov = np.cov(data_arr)
        inv_covmat = linalg.inv(cov)

        ## compute the rest of mahalnobis
        left_term = np.dot(x_minus_mu, inv_covmat)
        mahal = np.dot(left_term, x_minus_mu.T)
        mahald=mahal.diagonal()

        ## p=value 
        pvalue = 1 - chi2.cdf(mahald, 1)

        return mahald, pvalue

```
